```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA

class StockARIMAPredictor:
    def __init__(self, company_symbols, start_date, end_date):
        self.company_symbols = company_symbols
        self.start_date = start_date
        self.end_date = end_date

    def fetch_stock_data(self):
        """
        Fetches stock data for each company symbol from Yahoo Finance API
        """
        company_data = {}
        for symbol in self.company_symbols:
            company_data[symbol] = yf.download(symbol, self.start_date, self.end_date)
        return company_data

    def train_arima_models(self, company_data):
        """
        Trains ARIMA models for each company using its closing stock prices
        """
        arima_models = {}
        for symbol, data in company_data.items():
            train_data = data['Close']
            arima_model = ARIMA(train_data, order=(5, 1, 0))
            arima_fit = arima_model.fit()
            arima_models[symbol] = arima_fit
        return arima_models

    def make_predictions(self, arima_models, forecast_steps=None):
        """
        Makes predictions for future stock prices using trained ARIMA models
        """
        predictions = {}
        for symbol, arima_model in arima_models.items():
            if forecast_steps is None:
                forecast_steps = len(arima_model.data.endog)
            forecast = arima_model.get_forecast(steps=forecast_steps)
            forecast_values = forecast.predicted_mean.values
            predictions[symbol] = forecast_values
        return predictions

    def plot_predictions(self, company_data, predictions):
        """
        Plots actual stock prices alongside predicted stock prices for visual comparison
        """
        plt.figure(figsize=(16, 8))
        plt.title('ARIMA Model - Stock Price Predictions')
        plt.xlabel('Date')
        plt.ylabel('Close Price USD ($)')

        for symbol, data in company_data.items():
            plt.plot(data.index, data['Close'], label=f'{symbol} - Actual', linestyle='-')
            plt.plot(data.index, predictions[symbol], label=f'{symbol} - Predictions', linestyle='--')

        plt.legend()
        plt.show()

if __name__ == "__main__":
    company_symbols = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
    start_date = '2023-01-01'
    end_date = '2024-03-01'

    # Create an instance of the StockARIMAPredictor class
    stock_predictor = StockARIMAPredictor(company_symbols, start_date, end_date)

    # Fetch stock data
    company_data = stock_predictor.fetch_stock_data()

    # Train ARIMA models
    arima_models = stock_predictor.train_arima_models(company_data)

    # Make predictions
    predictions = stock_predictor.make_predictions(arima_models)

    # Plot predictions
    stock_predictor.plot_predictions(company_data, predictions)
```
